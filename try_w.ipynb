```python
import sys
## add the path to the sys path
sys.path.append('../stylegan2-ada-pytorch')

import torch
import pickle
import numpy as np
```

```python
with open('../stylegan2-ada-pytorch/stylegan2-afhqv2-512x512.pkl', 'rb') as f:
    G = pickle.load(f)['G_ema'].cuda()  # torch.nn.Module
z = torch.randn([1, G.z_dim]).cuda()    # latent codes
c = None                                # class labels (not used in this example)
img = G(z, c)                           # NCHW, float32, dynamic range [-1, +1]
```

```python
import matplotlib.pyplot as plt
plt.imshow(img[0].detach().cpu().permute(1, 2, 0).numpy())
```

```python
w = np.load('out_batch/projected_w_0.npz')['w']
w = torch.tensor(w).cuda()
print(w.shape)
```

```python
synth_image = G.synthesis(w, noise_mode='const')
synth_image = (synth_image + 1) * 0.5
```

```python
# Show the image
plt.imshow(synth_image[0].detach().cpu().permute(1, 2, 0).numpy())
plt.show()
```

```python
w_all = np.load('out_batch/projected_w_all.npz')['w']
print(w_all.shape)

fig, axes = plt.subplots(2, 6, figsize=(20, 6))
for i, ax in enumerate(axes.flatten()):
    w = w_all[i]
    w = torch.tensor(w).cuda()
    w = w.repeat([1, G.mapping.num_ws, 1])
    synth_image = G.synthesis(w, noise_mode='const')
    synth_image = (synth_image + 1) * 0.5
    ax.imshow(synth_image[0].detach().cpu().permute(1, 2, 0).numpy())
    ax.axis('off')
```
